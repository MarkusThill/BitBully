````python
# Only relevant for Google Colab:
import sys

IN_COLAB = "google.colab" in sys.modules

if IN_COLAB:
    from google.colab import output

    output.enable_custom_widget_manager()
````

````python
from bitbully import BitBully

# bitbully_agent: Connect4Agent = BitBully(opening_book="12-ply-dist", tie_break="center")

agents = {
    "BitBully-12ply": BitBully(opening_book="12-ply-dist", tie_break="random"),
    "peter": BitBully(opening_book="12-ply-dist", tie_break="random"),
}
````
